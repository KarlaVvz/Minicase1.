```python
import pandas as pd
import statsmodels.api as sm

# Assuming you have the data loaded in a pandas DataFrame called 'df'
df = pd.DataFrame({
    'mpg': [18.0, 15.0, 18.0, 16.0, 17.0],
    'cylinders': [8, 8, 8, 8, 8],
    'displacement': [307.0, 350.0, 318.0, 304.0, 302.0],
    'horsepower': [130.0, 165.0, 150.0, 150.0, 140.0],
    'weight': [3504, 3693, 3436, 3433, 3449],
    'acceleration': [12.0, 11.5, 11.0, 12.0, 10.5],
    'model_year': [70, 70, 70, 70, 70],
    'origin': ['usa', 'usa', 'usa', 'usa', 'usa'],
    'name': ['chevrolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino']
})

# Selecting the predictor variables (X) and target variable (y)
X = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']]
y = df['mpg']

# Adding a constant column to X for the intercept term
X = sm.add_constant(X)

# Creating an instance of the linear regression model
model = sm.OLS(y, X)

# Fitting the model to the data
model_fit = model.fit()

# Printing the model summary
print(model_fit.summary())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Assuming you have the data loaded in a pandas DataFrame called 'df'
df = pd.DataFrame({
    'mpg': [18.0, 15.0, 18.0, 16.0, 17.0],
    'cylinders': [8, 8, 8, 8, 8],
    'displacement': [307.0, 350.0, 318.0, 304.0, 302.0],
    'horsepower': [130.0, 165.0, 150.0, 150.0, 140.0],
    'weight': [3504, 3693, 3436, 3433, 3449],
    'acceleration': [12.0, 11.5, 11.0, 12.0, 10.5],
    'model_year': [70, 70, 70, 70, 70],
    'origin': ['usa', 'usa', 'usa', 'usa', 'usa'],
    'name': ['chevrolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino']
})

# Selecting the features and target variable
features = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']
target = 'mpg'

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Creating and training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Making predictions on the test data
predictions = model.predict(X_test)

print(predictions)
```
